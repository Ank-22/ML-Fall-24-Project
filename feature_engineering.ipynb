```python
infy_data['Open'] = pd.to_numeric(infy_data['Open'], errors='coerce')
infy_data['Close'] = pd.to_numeric(infy_data['Close'], errors='coerce')
infy_data['High'] = pd.to_numeric(infy_data['High'], errors='coerce')
infy_data['Low'] = pd.to_numeric(infy_data['Low'], errors='coerce')
infy_data['Volume'] = pd.to_numeric(infy_data['Volume'], errors='coerce')
infy_data['52_Week_High'] = pd.to_numeric(infy_data['52_Week_High'], errors='coerce')
infy_data["Is_52_week_high"] = (infy_data['Open'] >= infy_data['52_Week_High']).astype(int)
infy_data["Is_52_week_low"] = (infy_data['Open'] <= infy_data['52_Week_Low']).astype(int)
infy_data["Is_high"] = (infy_data['Open'] >= infy_data['High']).astype(int)
infy_data["Is_low"] = (infy_data['Open'] <= infy_data['Low']).astype(int)
infy_data['Maket_index'] = pd.to_numeric(market_index['Open'])
infy_data['Sector_index'] = pd.to_numeric(sector_index['Open'])
```

```python

# 1. Simple Moving Average (SMA)
infy_data['SMA_20'] = infy_data['Close'].rolling(window=20).mean()  # 20-day SMA
infy_data['SMA_50'] = infy_data['Close'].rolling(window=50).mean()  # 50-day SMA

# 2. Exponential Moving Average (EMA)
infy_data['EMA_20'] = infy_data['Close'].ewm(span=20, adjust=False).mean()  # 20-day EMA
infy_data['EMA_50'] = infy_data['Close'].ewm(span=50, adjust=False).mean()  # 50-day EMA

# 3. Bollinger Bands
# Middle band: 20-day SMA
# Upper band: 20-day SMA + 2 * 20-day rolling standard deviation
# Lower band: 20-day SMA - 2 * 20-day rolling standard deviation
infy_data['BB_upper'] = infy_data['SMA_20'] + 2 * infy_data['Close'].rolling(window=20).std()
infy_data['BB_lower'] = infy_data['SMA_20'] - 2 * infy_data['Close'].rolling(window=20).std()

# 4. Relative Strength Index (RSI)
delta = infy_data['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()

rs = gain / loss
infy_data['RSI'] = 100 - (100 / (1 + rs))

# 5. Moving Average Convergence Divergence (MACD)
# MACD = 12-day EMA - 26-day EMA
infy_data['MACD'] = infy_data['EMA_20'] - infy_data['EMA_50']

# Creating the traget and delta traget
infy_data['Target'] = infy_data['Close'].shift(-1)
infy_data['Delta_Target'] = infy_data['Target'] - infy_data['Close']

infy_data = infy_data.dropna(subset=['Target', 'Delta_Target'])
infy_data = infy_data.drop('Close', axis=1)
infy_data = infy_data.drop('Adj Close', axis=1)
```

```python
# Define the DataFrames and their names
dataframes = [market_index, sector_index, infy_data, ltim_data, tcs_data, persistent_data, wipro_data]
names = ["MARKET_DATA", "SECTOR_DATA", "INFY_DATA", "LTIM_DATA", "TCS_DATA", "PERSISTENT_DATA", "WIPRO_DATA"]

# Iterate through both DataFrames and their names
for df, name in zip(dataframes, names):
    # Refactoring the data
    df['Open'] = pd.to_numeric(df['Open'], errors='coerce')
    df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
    df['High'] = pd.to_numeric(df['High'], errors='coerce')
    df['Low'] = pd.to_numeric(df['Low'], errors='coerce')
    df['Volume'] = pd.to_numeric(df['Volume'], errors='coerce')
    df['52_Week_High'] = pd.to_numeric(df['52_Week_High'], errors='coerce')
    df["Is_52_week_high"] = (df['Open'] >= df['52_Week_High']).astype(int)
    df["Is_52_week_low"] = (df['Open'] <= df['52_Week_Low']).astype(int)
    df["Is_high"] = (df['Open'] >= df['High']).astype(int)
    df["Is_low"] = (df['Open'] <= df['Low']).astype(int)
    df['Maket_index'] = pd.to_numeric(market_index['Open'])
    df['Sector_index'] = pd.to_numeric(sector_index['Open'])

    
    # 1. Simple Moving Average (SMA)
    df['SMA_20'] = df['Close'].rolling(window=20).mean()  # 20-day SMA
    df['SMA_50'] = df['Close'].rolling(window=50).mean()  # 50-day SMA

    # 2. Exponential Moving Average (EMA)
    df['EMA_20'] = df['Close'].ewm(span=20, adjust=False).mean()  # 20-day EMA
    df['EMA_50'] = df['Close'].ewm(span=50, adjust=False).mean()  # 50-day EMA

    # 3. Bollinger Bands
    # Middle band: 20-day SMA
    # Upper band: 20-day SMA + 2 * 20-day rolling standard deviation
    # Lower band: 20-day SMA - 2 * 20-day rolling standard deviation
    df['BB_upper'] = df['SMA_20'] + 2 * df['Close'].rolling(window=20).std()
    df['BB_lower'] = df['SMA_20'] - 2 * df['Close'].rolling(window=20).std()

    # 4. Relative Strength Index (RSI)
    delta = df['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()

    rs = gain / loss
    df['RSI'] = 100 - (100 / (1 + rs))

    # 5. Moving Average Convergence Divergence (MACD)
    # MACD = 12-day EMA - 26-day EMA
    df['MACD'] = df['EMA_20'] - df['EMA_50']

    # Creating the traget and delta traget
    df['Target'] = df['Open'].shift(-1)
    df['Delta_Target'] = df['Target'] - df['Open']

    df.dropna(subset=['Target', 'Delta_Target'], inplace=True)
    df.drop('Close', axis=1, inplace=True)
    df.drop('Adj Close', axis=1,inplace=True)
    df.drop('High', axis=1,inplace=True)
    df.drop('Low', axis=1,inplace=True)
    df.to_csv(f'Dataset/{name}.csv', index=False)
```

# Some other stats that I didn't used

```python
# Signal Line: 9-day EMA of the MACD
infy_data['MACD_signal'] = infy_data['MACD'].ewm(span=9, adjust=False).mean()

# 6. Average True Range (ATR) - Volatility measure
high_low = infy_data['High'] - infy_data['Low']
high_close = (infy_data['High'] - infy_data['Close'].shift()).abs()
low_close = (infy_data['Low'] - infy_data['Close'].shift()).abs()

# True Range (TR)
true_range = pd.concat([high_low, high_close, low_close], axis=1)
infy_data['ATR'] = true_range.max(axis=1).rolling(window=14).mean()

# 7. On-Balance Volume (OBV) - Momentum Indicator
infy_data['OBV'] = np.where(infy_data['Close'] > infy_data['Close'].shift(), infy_data['Volume'], 
                            np.where(infy_data['Close'] < infy_data['Close'].shift(), -infy_data['Volume'], 0))

# 8. Stochastic Oscillator (%K and %D)
low_14 = infy_data['Low'].rolling(window=14).min()
high_14 = infy_data['High'].rolling(window=14).max()

# %K = (Close - 14-day Low) / (14-day High - 14-day Low) * 100
infy_data['%K'] = (infy_data['Close'] - low_14) / (high_14 - low_14) * 100

# %D = 3-day SMA of %K
infy_data['%D'] = infy_data['%K'].rolling(window=3).mean()

# 9. Commodity Channel Index (CCI)
typical_price = (infy_data['High'] + infy_data['Low'] + infy_data['Close']) / 3
sma_typical_price = typical_price.rolling(window=20).mean()
mean_deviation = (typical_price - sma_typical_price).abs().rolling(window=20).mean()
infy_data['CCI'] = (typical_price - sma_typical_price) / (0.015 * mean_deviation)
```
